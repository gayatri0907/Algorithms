```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import datetime as dt
```

```python
df=pd.read_csv('covid_19_india.csv',parse_dates=['Date'],dayfirst=True)
```

```python
df.head()
```

```python
df=df[['Date','State/UnionTerritory','Cured','Deaths','Confirmed']]
df.columns=['date','state','cured','deaths','confirmed']
```

```python
df.head()
```

```python
df.tail()
```

```python
today=df[df.date=='2020-06-22']
```

```python
today
```

```python
#sorting data w.r.t no. of confirmed cases
max_confirmed_cases=today.sort_values(by='confirmed',ascending=False)
max_confirmed_cases
```

```python
#states with max no. of confirmed cases
top_states_confirmed=max_confirmed_cases[0:5]
top_states_confirmed
```

```python
#plot for stats with top confirmed cases
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='confirmed',data=top_states_confirmed,hue='state')
plt.show()
```

```python
#plot for stats with max no. of deaths
max_death_cases=today.sort_values(by='deaths',ascending=False)
max_death_cases
```

```python
top_state_deaths=max_death_cases[0:5]
top_state_deaths
```

```python
#plot for states with max deaths
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='deaths',data=top_state_deaths,hue='state')
plt.show()
```

```python
#plot for stats with max no. of deaths
max_cured_cases=today.sort_values(by='cured',ascending=False)
max_cured_cases
```

```python
top_state_cured=max_cured_cases[0:5]
top_state_cured
```

```python
#plot for states with max cured
sns.set(rc={'figure.figsize':(15,10)})
sns.barplot(x='state',y='cured',data=top_state_cured,hue='state')
plt.show()
```
